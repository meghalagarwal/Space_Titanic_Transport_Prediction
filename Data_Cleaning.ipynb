```python
# Importing Libraries
import numpy as np
import pandas as pd
```

```python
# Creating a Data Preprocessing Class
class Data_preprocessing():
    # Importing Data
    def data_loading(self):
        data = pd.read_csv('/home/meghal/Personal/Personal Projects/Space Titanic/Data/train.csv')
        data = data.drop(['PassengerId', 'Name'], axis='columns')
        return data

    # Removing Null values with minimum threshold of 3
    def remove_null(self, data):
        df = data.dropna(thresh=len(data.columns)-2)

        return df

    # Filling Object Columns using Mode() operation
    def fill_object_columns(self, data):
        object_columns = data.select_dtypes(include='object').columns

        for col in object_columns:
            data[col] = data[col].fillna(data[col].mode()[0]) # Filling NULL Values of Object Datatypes Columns by taking the Mode.

        return data

    # Filling Floating Columns using Mean() & Median() operation
    def fill_float_columns(self, data):
        float_columns = data.select_dtypes(include='float64').columns

        for col in float_columns:
            if (col == 'Age'):
                data[col] = data[col].fillna(data[col].median()) # Filling NULL Values of Age by taking the Median.
            else:
                data[col] = round(data[col].fillna(data[col].mean()), ndigits=2) # Filling NULL Values of Amounts spent by Passengers by taking the Mean.

        return data
```

```python
df = Data_preprocessing().data_loading()
df
```

```python
# Checking the Data types of all the columns of the Dataframe
df.dtypes
```

```python
# Checking whether the data contains Null values.
df.isna().sum()
```

```python
# Removing Rows from the dataframe where there is more than 2 Null values.
df = Data_preprocessing().remove_null(df)
df.shape
```

```python
# Filling Null values seperately for Object Datatype Columns and Float Datatype Columns using Mean, Median & Mode
df = Data_preprocessing().fill_object_columns(df)
df = Data_preprocessing().fill_float_columns(df)
```

```python
# Checking whether the NULL Values are removed
df.isna().sum()
```

```python
# Storing the clean Dataframe into CSV File.
df.to_csv('/home/meghal/Personal/Personal Projects/Space Titanic/Data/cleaned_training_data.csv', index=False)
```
